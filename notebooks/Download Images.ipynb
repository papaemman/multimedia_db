**Step 2:** Interacting With Google Home Page

**How to write input text into the Google Images search bar.**

To do this you need to find the location of the search bar on the page. This is done by finding the search bar’s `Xpath.` 

1. To find the `Xpath` you need to open up the HTML of the web page which is done by clicking right anywhere on the page, and then pressing inspect.

2. Find the HTML code that corresponds to that Google search bar. To do this click the cursor icon on the top left of the box that just popped up.

3. Then click on the Google Images search box. When this is done the HTML on the right will update and a piece of HTML code will be highlighted in blue.

4. To get the Xpath of the search bar, you just need to right-click on the highlighted blue text and press “Copy Xpath”.

```python
# Setup the search box
box = driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input')
```

```python
# Define Keyword
keyword = 'Jeniffer Aniston'
```

```python
# Write keyword in the google search box and press enter
box.send_keys(keyword)
box.send_keys(Keys.ENTER)
```

**Step 3: Scrolling Down the Web Page**

Google Images is an **infinite scrolling web page**. This means as you keep scrolling down the web page new images will load in until there are no more images left.
Therefore, if we want to download hundreds of images we must first load them into the web page by scrolling down.

```python
# Will keep scrolling down the webpage until it cannot scroll no more
last_height = driver.execute_script('return document.body.scrollHeight')

while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    time.sleep(2)
    new_height = driver.execute_script('return document.body.scrollHeight')
    try:
        driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input').click()
        time.sleep(2)
    except:
        pass
    if new_height == last_height:
        break
    last_height = new_height
```

## Download (not Screenshot) images

```python
from selenium import webdriver
import time, requests

def search_google(search_query):
    browser = webdriver.Chrome("/home/user7/chromedriver_linux64/chromedriver")
    search_url = f"https://www.google.com/search?site=&tbm=isch&source=hp&biw=1873&bih=990&q={search_query}"
    images_url = []

    # open browser and begin search
    browser.get(search_url)
    elements = browser.find_elements_by_class_name('rg_i')

    count = 0
    
    for e in elements:
        
        try:
            # get images source url
            e.click()
            time.sleep(1)
            element = browser.find_elements_by_class_name('v4dQwb')

            # Google image web site logic
            if count == 0:
                big_img = element[0].find_element_by_class_name('n3VNCb')
            else:
                big_img = element[1].find_element_by_class_name('n3VNCb')

            images_url.append(big_img.get_attribute("src"))


            # write image to file
            reponse = requests.get(images_url[count])
            if reponse.status_code == 200:
                with open(f"../data/actors/{search_query}/{search_query}_{count+1}.jpg","wb") as file:
                    file.write(reponse.content)
        except:
            pass
        
        count += 1
        print("Count:", count, end="\n")
        # Stop get and save after N
        if count >= 50:
            break

    return images_url
```
