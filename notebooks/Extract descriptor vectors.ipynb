# Extract descriptor vectors from images

Tutorial:
https://www.digitalocean.com/community/tutorials/how-to-detect-and-extract-faces-from-an-image-with-opencv-and-python

**Function**

1. Identify all faces in the photo
2. Extract face encodings for every face in this image
3. Return {file_path, face_encoding, main_face} for every face

```python
# Import face_recognition module (simple API)
import face_recognition
```

```python
actor = "Robert De Niro"
```

```python
image_path = f"../data/actors/{actor}/{actor}_18.jpg"
```

```python
from IPython.display import Image, display
display(Image(filename=image_path) )
```

```python
def extract_photo_descriptors(actor, image_path):
    
    # Create an encoding of facial features that can be compared to other faces
    picture = face_recognition.load_image_file(file=image_path)
    
    res = []
    for face_encoding in face_recognition.face_encodings(picture):
        res.append({"face_encoding":face_encoding,
                    "image_path": image_path, 
                    "actor":image_path.split("/")[2]})
    
    return res
```

```python
extract_photo_descriptors(actor, image_path)
```
